```python
def limpiar_fecha(s):
    
    x = s.split('.')
    lista=[]
    
    for i in x:
        
        lista.append(i)
            
    return ('.'.join(lista[0:3]))
```

```python
def quitar_fechas(a):
    
    for i in a:
        
        if len(a)!=10:
            
            return 'unknown'
        
        if len(a)==10:
            
            return a
```

```python
def contar_fechas(c):
    
    lista=[]
    
    for i in c:
        
        if i[8:10]=='00':
            
            lista.append(i)
    
    return len(lista)
```
